```python
from sklearn.model_selection import train_test_split      
```

```python
from sklearn.linear_model import LogisticRegression           
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix          
```

```python
data=data.drop(columns='Name')      
```

```python
new_data=pd.get_dummies(data)   #particular column has many sub type then split 
new_data                         #and use 0 &1 so we can use Logistic regression easily
```

```python
column_list=new_data.columns
column_list
```

```python
x=new_data.drop(['Probablity'],axis=1).values
x
```

```python
y=new_data['Probablity'].values                 #storing output variabe in y
print(y)
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.4,random_state=0) 
```

```python
logistic=LogisticRegression()   #model 
```

```python
test_x
```

```python
test_y
```

```python
test_x
```

```python
logistic.fit(train_x,train_y)  #fit training data
```

```python
prediction=logistic.predict(test_x)       #prediction from the test data 
prediction
```

```python
confusion_matrix=confusion_matrix(test_y, prediction)     #confusion matrix   
confusion_matrix
```

```python
accuracy_score=accuracy_score(test_y, prediction)        #calculating the accuracy
print(accuracy_score)                                    #This is the accuracy it can be mor accurate if we romove some more 
                                                         
```

```python
print("No of misclassified value :",(test_y!=prediction).sum())
```
